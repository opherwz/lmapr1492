```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
```

```python
with MPRester("P2FTs1l6Xaon8AOI") as m: 
    struct = m.get_structure_by_material_id("mp-567259")
    
xrdcalc = XRDCalculator(wavelength = 'CuKa', symprec = 0, debye_waller_factors = None)
xrdpattern = xrdcalc.get_pattern(structure = struct, scaled = True, two_theta_range = (0,90))

xrdcalc.show_plot(struct, two_theta_range = (0,25)) #pour ne plotter que les 3 premiers pics
for i in range(0,3) :
    
    print("Pic de diffraction n°", i+1 , " : 2 \u0398 = ", xrdpattern.x[i], 'degrés')
    print("                       intensité ([0,100]) : ", xrdpattern.y[i]) 
    print("                       hkl : ", xrdpattern.hkls[i][0]['hkl'])
    print("                       multiplicité du pic : ", xrdpattern.hkls[i][0]['multiplicity'])
    print('\n')
```
